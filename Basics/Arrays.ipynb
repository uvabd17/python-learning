# 1) Find the largest element in the array:

```python
def largest_element(arr):
    n=len(arr)
    max1=0
    for i in range(n):
        if arr[i]>max1:
            max1=arr[i]
    return max1
arr=list(map(int, input().split(",")))
print(largest_element(arr))
```

# 3) Check if array is sorted and rotated

```python
def check_sorted(arr,n):
    cnt=0
    for i in range(n):
        if arr[i-1]>arr[i]:
            cnt+=1
    if cnt<=1:
        return True
    else:
        return False
arr=list(map(int, input().split(",")))
n=len(arr)
check_sorted(arr,n)
```

# 4) Remove Duplicates from the sorted array

```python
# Brute Force Approach 
def remove_dups(arr, n):
    arr2=set()
    for i in range(n):
        arr2.add(arr[i])
    size=0
    for val in arr2:
        arr[i]=val
        size+=1
    return size
arr=[1,1,2,2,2,3,3]
n=len(arr)
remove_dups(arr,n)
```

```python
def rem_dups(arr,n):
    i=0
    for j in range(1,n):
        if arr[j]!=arr[i]:
            arr[i+1]=arr[j]
            i+=1
    return i+1
arr=[1,1,2,2,2,3,3,3]
n=len(arr)
rem_dups(arr,n)
```

# 5) Left rotate the array by one place

```python
def rotate_left_one(arr):
    temp=arr[0]
    n=len(arr)
    for i in range(1,n):
        arr[i-1]=arr[i]
    arr[n-1]=temp
    pass
arr=[1,2,3,4,5]
rotate_left_one(arr)
print(arr)
```
