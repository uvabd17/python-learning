# 1) Find the largest element in the array:

```python
def largest_element(arr):
    n=len(arr)
    max1=0
    for i in range(n):
        if arr[i]>max1:
            max1=arr[i]
    return max1
arr=list(map(int, input().split(",")))
print(largest_element(arr))
```

# 3) Check if array is sorted and rotated

```python
def check_sorted(arr,n):
    cnt=0
    for i in range(n):
        if arr[i-1]>arr[i]:
            cnt+=1
    if cnt<=1:
        return True
    else:
        return False
arr=list(map(int, input().split(",")))
n=len(arr)
check_sorted(arr,n)
```

# 4) Remove Duplicates from the sorted array

```python
# Brute Force Approach 
def remove_dups(arr, n):
    arr2=set(arr)
    for i in range(n):
        arr2.add(arr[i])
    size=0
    for val in arr2:
        arr[size]=val
        size+=1
    return size
arr=[1,1,2,2,2,3,3]
n=len(arr)
remove_dups(arr,n)
```

# 6) Left rotate by k places 

```python
# Brute Force Approach
def left_rotate_k(arr, k):
    n=len(arr)
    k=k%n # why: because f n=7 and k=7, the array is same, and if k=8, n=7, then only 1 rotation is needed
    temp=[arr[i] for i in range(0,k)]
    #shifting
    for i in range(k,n):
        arr[i-k]=arr[i]
    #putback   
    for i in range(n-k,n):
        arr[i]=temp[i-(n-k)]
arr=[1,2,3,4,5,6,7,8]
left_rotate_k(arr,19)
print(arr)
```

# 7) Move Zeroes to end

```python
# Brute Force:
def move0_to_end(arr):
    temp=[]
    #step 1: temp append non zeros
    for i in range(0, len(arr)):
        if arr[i]!=0:
            temp.append(arr[i])
    nzero=len(temp)
    # step2: change the arr values using temp
    for i in range(0, len(temp)):
        arr[i]=temp[i]
    # step 3: assign zeroes to the end
    for i in range(nzero, len(arr)):
        arr[i]=0
arr=[1,2,0,4,5,3,2,0,7,8,0,0,5]
move0_to_end(arr)
print(arr)   
```

TC: O(n) + O(x) + O(n-x) --> O(n)
SC: O(n) temp 
<br>
No better solution , directly to optimal

```python
# OPTIMAL SOLUTION
def  move0_to_end_opt(arr):
    j=-1
    n=len(arr)
    for i in range(0,n):
        if arr[i]==0:
            j=i
            break
    if j==-1 : return arr
    for i in range(j+1,n):
        if arr[i]!=0:
            arr[i],arr[j]=arr[j],arr[i]
            j+=1
    return arr
arr=[1,2,0,4,5,3,2,0,7,8,0,0,5]
print(move0_to_end_opt(arr))
```

# 8) Linear Search

```python
def linear_search(arr, num):
    for i in range(len(arr)):
        if arr[i]==num:
            return i
    return -1
arr=[6,7,8,4,1]
linear_search(arr, 7)
```

Worst Case(if all are unique): TC: O(n1logn + n2logn) + O(n1+n2) <br>
SC: O(n1+n2) + O(n1+n2) {this another n1+n2 is to return the answer, not to solve the prob}

```python
# Optimal Approach
def union_opt(a,b):
    i=0
    j=0
    n1=len(a)
    n2=len(b)
    union=[]
    while(i<n1 and j<n2):
        if a[i]<=b[j]:
            if len(union)==0 or union[-1]!=a[i]:
                union.append(a[i])
            i+=1
        else:
            if len(union)==0 or union[-1]!=b[j]:
                union.append(b[j]) 
            j+=1

    while j<n2:
        if len(union)==0 or union[-1]!=b[j]:
            union.append(b[j]) 
        j+=1
    while i<n1:
        if len(union)==0 or union[-1]!=a[i]:
            union.append(a[i])
        i+=1
    return union
a=[1,2,3,4,5,5,6,7]
b=[2,2,3,5,6,7,8]
union_opt(a,b)
```

TC: O(n+m)
<br>
SC: only to return or output O(n+m) or O(n1+n2) if all are unique, worst case

# 9) Intersection Of two "SORTED" arrays

```python
# Brute Force Approach:
def intersection(a, b):
    vis=[0]*len(b)
    result=[]
    for i in range(len(a)):
        for j in range(len(b)):
            if a[i]==b[j] and vis[j]==0:
                result.append(a[i])
                vis[j]=1
                break
            if b[j]>a[i] : break
    return result
a=[1,2,3,4,5,5,6,7]
b=[2,2,3,5,6,7,8]
intersection(a,b)
```

TC: O(n1*n2)
<br>
SC: O(n2) the vis array is same size as b , may change how the algo has been written

```python
# OPTIMAL APPROACH:
# Two pointer approach
def intersection2(a,b):
    i=0
    j=0
    n1=len(a)
    n2=len(b)
    ans=[]
    while i<n1 and j<n2:
        if a[i]<b[j]:
            i+=1
        elif b[j]>a[i]:
            j+=1
        else:
            ans.append(a[i])
            i+=1
            j+=1
    return ans
a=[1,2,3,4,5,5,6,7]
b=[2,2,3,5,6,7,8]
intersection2(a,b)
```

TC: (worst case) : O(n1+n2)
<br>
SC: O(1) , no extra space
