# 1) Selection Sort

<h3>Select Min and swap with the first place

```python
def selection_sort(arr):
    n=len(arr)
    for i in range(n-1):
        mini=i
        for j in range(i,n):
            if arr[j]<arr[mini]:
                mini=j
        temp=arr[mini]
        arr[mini]=arr[i]
        arr[i]=temp

arr= list(map(int, input().split(",")))
selection_sort(arr)
print(arr)
```

Time Complexity: O(n**2/2 + n/2) ( sum of n numbers) which is rounded off to O(n**2)

# 2) Bubble Sort 

<h3> Pushes the Max to the last. Opposite of selection sort. ADJACENT SWAPPING</h3>

```python
def bubble_sort(arr):
    '''
    Remember the last one has nothing to check with, 
    so causes runtime error , hence till 'i-1', 
    means till i-2 because of range function
    '''
    n=len(arr)
    didSwap=0
    for i in range(n,-1,-1):
        for j in range(0,i-1): 
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                didSwap=1
        if didSwap==0:
            break
        print('runs')
arr= list(map(int, input().split(",")))
bubble_sort(arr)
print(arr)
```

So, TIme Complexity: O(n**2) --> worst and avg
but, so what's the best? when the array is sorted already , means the array itself is already in the ascending order, so no swaps needed. for that we can optimize it. the code is in comments. uncomment to check the optimized one. 

Best TC: O(n) 

# 3) Insertion Sort  

Takes an element and places it in its correct order

```python
def insertion_sort(arr):
    n=len(arr)
    for i in range(n):
        j=i
        while j>0 and arr[j-1]>arr[j]:
            arr[j],arr[j-1]=arr[j-1],arr[j]
            j-=1
arr= list(map(int, input().split(",")))
insertion_sort(arr)
print(arr)
```

Best case when the array is already sorted, the inner loop doesn't run at all. that means the outer loop runs, and that too just once. so: O(n)
but worst and avg would be O(n**2)
