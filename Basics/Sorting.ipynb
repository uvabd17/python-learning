# 1) Selection Sort

<h3>Select Min and swap with the first place

```python
def selection_sort(arr):
    n=len(arr)
    for i in range(n-1):
        mini=i
        for j in range(i,n):
            if arr[j]<arr[mini]:
                mini=j
        temp=arr[mini]
        arr[mini]=arr[i]
        arr[i]=temp

arr= list(map(int, input().split(",")))
selection_sort(arr)
print(arr)
```

Time Complexity: O(n**2/2 + n/2) ( sum of n numbers) which is rounded off to O(n**2)

# 2) Bubble Sort 

<h3> Pushes the Max to the last. Opposite of selection sort. ADJACENT SWAPPING</h3>

```python
def bubble_sort(arr):
    '''
    Remember the last one has nothing to check with, 
    so causes runtime error , hence till 'i-1', 
    means till i-2 because of range function
    '''
    n=len(arr)
    didSwap=0
    for i in range(n,-1,-1):
        for j in range(0,i-1): 
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                didSwap=1
        if didSwap==0:
            break
        print('runs')
arr= list(map(int, input().split(",")))
bubble_sort(arr)
print(arr)
```

So, TIme Complexity: O(n**2) --> worst and avg
but, so what's the best? when the array is sorted already , means the array itself is already in the ascending order, so no swaps needed. for that we can optimize it. the code is in comments. uncomment to check the optimized one. 

Best TC: O(n) 

# 3) Insertion Sort  

Takes an element and places it in its correct order

```python
def insertion_sort(arr):
    n=len(arr)
    for i in range(n):
        j=i
        while j>0 and arr[j-1]>arr[j]:
            arr[j],arr[j-1]=arr[j-1],arr[j]
            j-=1
arr= list(map(int, input().split(",")))
insertion_sort(arr)
print(arr)
```

Best case when the array is already sorted, the inner loop doesn't run at all. that means the outer loop runs, and that too just once. so: O(n)
but worst and avg would be O(n**2)

# 3) MERGE SORT

Divide and Merge : Recursion is involved. Very important. 

```python
def mergesort(arr, low, high):
    if low>=high: return
    mid=(low+high)//2
    mergesort(arr, low, mid)
    mergesort(arr, mid+1, high)
    merge(arr, low, mid, high)
    return arr

def merge(arr, low, mid, high):
    '''
    temporary list to store the sorted elements, append them
    one by one.
    left and right are two pointers which move across the divided
    lists/arrays. 
    '''
    temp=[]
    left=low
    right=mid+1
    while(left<=mid and right<=high):
        if arr[left]<arr[right]:
            temp.append(arr[left])
            left+=1
        else:
            temp.append(arr[right])
            right+=1
    '''
    Important Point : 
    Till above the comparision ends if one pointer crosses 
    their end(mid/high). So, for those remaining elements, we
    add the below code. which is important to know. 
    '''
    while left<=mid:
        temp.append(arr[left])
        left+=1
    while right<=high:
        temp.append(arr[right])
        right+=1

    # Below, stores all the sorted temp elements into the original array/list
    for i in range(low,high+1):
        arr[i]=temp[i-low]

    return arr

arr=list(map(int,input().split(",")))
low=0
high=len(arr)-1
mergesort(arr,low,high)
```

# 4) Recursive Bubble Sort

Now, how shall we approach this. The idea of bubble sort is , we push the max element to the last just how a bubble is blown from a bubble ring. 

```python
def recursive_bubble_sort(arr, n):
    '''
    forgot the point that for every sort, the max element goes to 
    the last, and we sort the next part only till before that max element.
    hence the n-1 in the recursion call.
    didSwap here is important, because it is also a base case.
    n==1 is a case where the array to sort left is only one element
    (happens in the worst case)
    '''
    didSwap=0
    if n==1: return 
    for i in range(n-1):
        if arr[i]>arr[i+1]:
            arr[i],arr[i+1]=arr[i+1],arr[i]
            didSwap=1

    if didSwap==0: return 
        
    recursive_bubble_sort(arr,n-1)
    
arr=list(map(int, input().split(",")))
n=len(arr)
recursive_bubble_sort(arr,n)
print(arr)
```

# 5) Recursive Insertion Sort

Think this through carefully. 

```python
def ris(arr,i,n):
    if i==n:
        return
    j=i
    while (j>0 and arr[j-1]>arr[j]):
        arr[j],arr[j-1]=arr[j-1],arr[j]
        j-=1
    ris(arr,i+1,n)
arr=[3,2,4,1,2]
ris(arr,0,len(arr))
print(arr)
```

# 6) QUICK SORT

This is a bit complex. Divide and Conquer

```python
def partition(arr, low, high):
    pivot=arr[low]
    i=low
    j=high
    while(i<j):
        '''
        For descending, change the comparisions. that's all. 
        '''
        while(i<=high-1 and arr[i]<pivot):
            i+=1
        while(j>=low+1 and arr[j]>=pivot):
            j-=1
        if i<j : arr[i],arr[j]=arr[j],arr[i]
    arr[low],arr[j]=arr[j],arr[low]
    return j

def qs(arr, low, high):
    if low<high:
        pIndex=partition(arr,low,high)
        qs(arr, low, pIndex-1)
        qs(arr, pIndex+1, high)

arr=[4,6,2,5,7,9,1,3]
qs(arr,0,len(arr)-1)
print(arr)    
```
