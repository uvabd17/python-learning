2. **Functions**-organizing your code better
Function: A block of code designed to perform a specific task without duplication of code.
   

```python
#Example
# pay = no.of hours * pay per hour
# 12% tax deductions/ 12 % tax bracket
def get_pay(num_hours):
    pay_preTax=num_hours*15
    pay_afterTax=pay_preTax*(1-0.12)
    return pay_afterTax
pay_full=get_pay(40)
print(pay_full)

```

```python
#scope : var inside a body has its scope of access or usage only inside that same body
print(pay_afterTax)
```

```python
#multiple arguments
def get_pay_with_more_inputs(num_hours, hourly_wage, tax_bracket):
    pay_preTax=num_hours*hourly_wage
    pay_afterTax=pay_preTax*(1-tax_bracket)
    return pay_afterTax
higher_pay_afterTax=get_pay_with_more_inputs(56,40,.32)
print(higher_pay_afterTax)    
```

```python
def get_cost(sqft_walls, sqft_ceiling, sqft_per_gallon, cost_per_gallon):
    #total area/area covered by one gallon of paint = no.of gallons req
    cost=cost_per_gallon*((sqft_walls+sqft_per_gallon)/sqft_per_gallon)#doesn't round off, '/' division for floating point 
    #'//' division for int values only
    return cost
get_cost(40,50,3,20)
    
```

```python
project_cost=get_cost(432,144,400,15)
print(project_cost)
```

```python
#math.ceil() #valid to use function inside a function
import math
def get_actual_cost(walls, ceiling, sqft_per_gallon, cost_per_gallon):
    cost=cost_per_gallon*(math.ceil((walls+ceiling)/sqft_per_gallon))
    return cost
get_actual_cost(456, 78,23,16)
```

In this question, you'll use this list comprehension to define a function `percentage_liked()` that takes one argument as input:
- `ratings`: list of ratings that people gave to a movie, where each rating is a number between 1-5, inclusive

We say someone liked the movie, if they gave a rating of either 4 or 5.  Your function should return the percentage of people who liked the movie.

For instance, if we supply a value of `[1, 2, 3, 4, 5, 4, 5, 1]`, then 50% (4/8) of the people liked the movie, and the function should return `0.5`.

Part of the function has already been completed for you.  You need only use `list_liked` to calculate `percentage_liked`.

```python
def percentage_liked(ratings):
    list_liked=[i>=4 for i in ratings]
    percentage_liked=sum(list_liked)/len(list_liked)
    print("sum of booleans here gives the no.of True values (0+0+0+1+1+1+1+0)")
    return percentage_liked
percentage_liked([1,2,3,4,5,4,5,1])
```

```python
def least_difference(a,b,c):
    """ Return the smallest difference between any two numbers among 
    a,b and c. 
    """
    diff1=abs(a-b)
    diff2=abs(b-c)
    diff3=abs(c-a)
    return min(diff1, diff2, diff3)
```
