```python
#Brute Forcing the counting of numbers in an array:
def counter(number, arr):
    cnt=0
    for i in range(len(arr)):
        if number==arr[i]:
            cnt+=1
    return cnt
counter(2, [1,2,1,3,2])
```

<h2> Number Hashing</h2>

```python
# using HASH array or Frequency Array
class Solution:
    #pre compute
    def __init__(self, arr):
        self.hash1={}
        for num in arr:
            self.hash1[num]=self.hash1.get(num,0)+1
    #fetch
    def count(self, num):
        return self.hash1.get(num,0)
n=int(input())
arr=list(map(int, input().split(" ")))
assert len(arr)==n
obj=Solution(arr)
t=int(input())
while t>0:
    num=int(input())
    print(obj.count(num))
    t-=1
        
        
```

<h2>Character Hashing: 

```python
class CharacterHashing:
    #pre-compute
    def __init__(self, s):
        self.freq={}
        for char in s:
            self.freq[char]=self.freq.get(char,0)+1

    def count(self, char):
        return self.freq.get(char,0)
s=input()
obj=CharacterHashing(s)
print(obj.count('a'))
```
