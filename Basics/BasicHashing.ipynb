```python
#Brute Forcing the counting of numbers in an array:
def counter(number, arr):
    cnt=0
    for i in range(len(arr)):
        if number==arr[i]:
            cnt+=1
    return cnt
counter(2, [1,2,1,3,2])
```

This approach is very bad even though it works for small values , small sized array , and small numbers. but what if it had to say the count mulitple times?  TIME COMLEXITY: O( Q (no.of numbers) x N (the loop runs for N times)) == O(QxN)

<h1>HASHING: Pre Storing and Fetching When required in naive terms

<h2> Number Hashing</h2>

```python
# using HASH array or Frequency Array
class Solution:
    #pre compute
    def __init__(self, arr):
        self.hash1={}
        for num in arr:
            self.hash1[num]=self.hash1.get(num,0)+1
    #fetch
    def count(self, num):
        return self.hash1.get(num,0)
n=int(input())
arr=list(map(int, input().split(" ")))
assert len(arr)==n
obj=Solution(arr)
t=int(input())
while t>0:
    num=int(input())
    print(obj.count(num))
    t-=1
        
        
```

<h2>Character Hashing: 

```python
class CharacterHashing:
    #pre-compute
    def __init__(self, s):
        self.freq={}
        for char in s:
            self.freq[char]=self.freq.get(char,0)+1

    def count(self, char):
        return self.freq.get(char,0)
s=input()
obj=CharacterHashing(s)
t=int(input())
while t>0:
    print(obj.count(input()))
    t-=1
#TC: O(n+q) n: no.of loop run + q is the no.of times the char on which loop ran
```

<h2> 3) Highest Occuring Element in an Array </h2>

```python
class Solution:
    def maxFreq(self, nums: list[int], k:int)->int:
        arr=sorted(nums)
        max_count=1
        for i in range(len(arr)-1,-1,-1):
            chk=0
            cnt=0
            for j in range(i-1,-1,-1):
                chk=chk+(arr[i]-arr[j])
                cnt+=1
                if chk<=k:
                    max_count=max(max_count,cnt+1)
                else:
                    continue
        return max_count
obj=Solution()
nums=[9926,9960,10000,9992,9917,9986,9934,9985,9977,9950,9922,9913,9971,9978,9984,9959,9934,9948,9918,9916,9967,9965,9985,9977,9988,9983,9900,9945,9913,9966,9968,9986,9939,9914,9980,9957,9921,9927,9917,9972,9974,9953,9984,9912,9975,9920,9966,9932,9921,9904,9928,9959,9993,9937,9934,9974,9937,9964,9922,9963,9991,9930,9944,9930,9982,9980,9967,9904,9955,9947,9924,9973,9997,9950,9905,9924,9990,9947,9953,9924,9977,9938,9951,9982,9932,9926,9928,9912,9917,9929,9924,9921,9987,9910,9927,9921,9929,9937,9919,9995,9949,9953]
k=3044
obj.maxFreq(nums,k)
```
