 # Recursion
-------------- <br>
Two Important things: 
1. A function calls itself
2. until base condition meets. <br>
**BASE CONDITION** is important, can be one or multiple. <br>
**STACK OVERFLOW**:<br>
until the funtion self call inside the funtion isn't executed, that function isn't considered to be executed completely, so that stays in the stack memory. If recursion is infinite, the stack gets completely filled and overflows, causing segmentation error or stack overflow.

```python
cnt=0
def f():
    global cnt
    if cnt==4:
        return
    print(cnt)
    cnt+=1
    f()
    
f()
```

#  Basic Problems - Recursion

1. Print a name N times

```python
def names(name, N):
    if N==0:
        return
    print(name)
    N-=1
    names(name, N)
names('uvabd17' , 5)
    
```
