**STACK OVERFLOW**:<br>
until the funtion self call inside the funtion isn't executed, that function isn't considered to be executed completely, so that stays in the stack memory. If recursion is infinite, the stack gets completely filled and overflows, causing segmentation error or stack overflow.

# 1) Sum of First N numbers

Two ways : <br>
1. Parameter
2. Funtion

```python
# 1) PARAMETERIZED
def sum1(i, sum):
    if i<1:
        print(sum)
        return
    sum1(i-1, sum+i)
sum1(10,0)
```

```python
# 2) FUNCTIONAL
#Only Use funtion to get the answers, no dependency on the parameters
def sum2(n):
    if n==0:
        return 0
    return n+sum2(n-1)
sum2(10) 
```

 # 2) Factorial of a Number

```python
# Parameterized
def fact1(i, fact):
    if i==1:
        print(fact)
        return
    fact1(i-1, fact*i)
fact1(5,1)
```

```python
# functional
def fact2(n):
    if n==0:
        return 1
    return n*fact2(n-1)
fact2(5) # TC: O(n) (no.of calls) SC: O(n) (n calls are waiting)
```

# 2) Reverse an Array

```python
# Two Pointers:
def reverse(arr, p1,p2):
    if p1>=p2:
        return 
    arr[p1],arr[p2]=arr[p2],arr[p1]
    reverse(arr, p1+1,p2-1)
reverse([1,2,3,4,2,6],0,5)
print(arr)
```

```python
# One variable/pointer/index:
def rev_arr(arr,i,n):
    if i>=n/2:
        return
    arr[i], arr[n-i-1]=arr[n-i-1],arr[i]
    rev_arr(arr, i+1,n)
arr=[1,2,3,4,2,6]
rev_arr(arr,0,len(arr))
print(arr)
```

# 3) Check if a string is a palidrome

```python
# python classic:
def pal_check(str):
    return str==str[::-1]
pal_check('madam')
```

```python
# striver bhai:
# use the reverse array method, but here don't have to swap
class Solution:
    def palindrome_check(self, str1:str, i)->bool:
        if i>=len(str1)/2:
            return True
        if str1[i]!=str1[len(str1)-i-1]:
            return False
        return self.palindrome_check(str1, i+1)
obj=Solution()
str1="madam"
obj.palindrome_check(str1,0)
```

```python
import re
class Solution:
    def isPalindrome(self, s: str) -> bool:
        str1="".join(s.split(" "))
        str1=re.sub(r'[^A-Za-z0-9\s]','',str1)
        return str1.lower()==str1[::-1].lower()

obj=Solution()
s="A man, a plan, a canal: Panama"
obj.isPalindrome(s)
```

# Multiple Recursion Calls

1) Fibonacci Number

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
fib(10) # TC: 2**n
```

TC : No.of recursion calls for f(4) is 9 one direction calls, way beyond 4 calls, so for every func call we call 2 function calls. 
1--> 2--> 2--> 2 --> 2 ... (2*2*2*..) = 2**n
