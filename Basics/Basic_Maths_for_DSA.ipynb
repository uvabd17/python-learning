# 1) Extract all the digits

```python
def extract_digits(num):
    digit=[]
    while num>0:
        digit.append(num%10)
        num=num//10
    print(digit[::-1])
    print(len(digit))
    pass

n=int(input("Enter a Number:"))
extract_digits(n)
```

# 2) Count the Digits

```python
def count_digits(num):
    num=abs(num)
    if num==0:
        return 1
    count=0
    while num>0:
        count = count + 1
        num=num//10
    return count
```

```python
count_digits(-9877)
```

# 3) Reverse a Number

```python
class Solution:
    def reverse(self, x:int)->int:
        a=abs(x)
        revNum=0
        while a>0:
            ld=a%10
            a=a//10
            revNum=(revNum *10)+ld
        if x<0:
            return -revNum
        else:
            return revNum
obj=Solution()
```

# 4) Palindrome Number

```python
n=int(input("Enter:"))
obj.reverse(n)
if obj.reverse(n)==n:
    print(True)
else: 
    print(False)
 
```

# 5) Armstrong Numbers

```python
class Solution:
    def armstrong(self, num: int)-> int:
        sum=0
        a=num
        while a>0:
            ld=a%10
            sum = sum + ld**3
            a = a//10
        return sum==num
obj=Solution()
obj.armstrong(371)
```

# 6) Print All Divisors

```python
# TC : O(n)
def all_factors(n):
    return [i for i in range(1, n+1) if n%i==0 ]
all_factors(36)    
```

```python
# TC: O(sqrt(N)) + O (n (no.of factors) * log n )
import math
def all_factors2(n):
    a=[]
    for i in range( 1, math.isqrt(n)+1): # also i*i<=n works 
        if n%i==0:
            a.append(i)
        if (n//i)!=i:
            a.append(n//i)
    return sorted(a) # takes O(nlogn) n is no.of factors
all_factors2(36)
```

# 7) Prime Check

```python
# Factors play a role, 2 factors only, then it's a prime 
import math
def prime_check(n):
    cnt=0
    for i in range( 1 , math.isqrt(n)+1):
        if n%i==0:
            cnt+=1
        if n//i!=i:
            cnt+=1
    return cnt==2
prime_check(1) # TC: O(sqrt(n))
```
