# 1) Extract all the digits

```python
def extract_digits(num):
    digit=[]
    while num>0:
        digit.append(num%10)
        num=num//10
    print(digit[::-1])
    print(len(digit))
    pass

n=int(input("Enter a Number:"))
extract_digits(n)
```

# 2) Count the Digits

```python
def count_digits(num):
    num=abs(num)
    if num==0:
        return 1
    count=0
    while num>0:
        count = count + 1
        num=num//10
    return count
```

```python
count_digits(-9877)
```

# 3) Reverse a Number

```python
class Solution:
    def reverse(self, x:int)->int:
        a=abs(x)
        revNum=0
        while a>0:
            ld=a%10
            a=a//10
            revNum=(revNum *10)+ld
        if x<0:
            return -revNum
        else:
            return revNum
obj=Solution()
```

# 4) Palindrome Number

```python
n=int(input("Enter:"))
obj.reverse(n)
if obj.reverse(n)==n:
    print(True)
else: 
    print(False)
 
```
